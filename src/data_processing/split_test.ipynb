# This Jupyter Notebook is for testing the split data

```python
import h5py
import numpy as np

with h5py.File("../../data/lotka_volterra_train.h5", "r") as f:
    # Access the full dataset
    trajectories = f["trajectories"][:]
    time_points = f["time"][:]

    print("Shape of trajectories:", trajectories.shape)
    print("Shape of time points:", time_points.shape)

    # Access a single trajectory
    system_id = 0 # First system
    
    prey = trajectories[0, :, 0]
    predator = trajectories[0, :, 1]
    times = time_points[:]

    # visualize the data
    import matplotlib.pyplot as plt
    plt.plot(times, prey, label="Prey")
    plt.plot(times, predator, label="Predator")
    plt.xlabel("Time")
    plt.ylabel("Population")
    plt.legend()
    plt.show()
```
